```python
import cfgrib
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")

from baselines.data_processor import DataProcessor
from baselines.grad_boost.grad_booster import GradBooster
from time import time 
```

```python
processor = DataProcessor()
data, feature_list = DataProcessor.load_data()
s = 5
fh = 1

X, y = processor.preprocess(s, fh, use_neighbours=False)
X_train, X_test, y_train, y_test = processor.train_val_test_split(X, y, split_type=2) # train 2020; test 2022
```

```python
for boo in ["lgb", "xgb"]: #, "cat"]: # , "ada"]: # ada is extremely slow
    start = time()
    booster = GradBooster(X.shape, fh, feature_list, boo)
    booster.train(X_train, y_train)
    _ = booster.predict_and_evaluate(X_test, y_test, max_samples=1)
    print()
    stop = time()
    print(f'\n{boo}: {stop-start} [S]\n')    
```
